Aplica un descuento.
Descripción: Estamos de oferta en una tienda y queremos aplicar un descuento a todos los productos. Pasa una lista de precios a la función. Si el valor es inferior a 10, aplícale un 5% de descuento. Si está entre 10 y 40, aplícale un 15%, si vale más, aplícale un 30%. Devuelve la lista de valores tras aplicar al descuento.


```python
def descuento(productos):
    descuento_aplicado=[]
    t1=0.05
    t2=0.15
    t3=0.3

    for p in productos:
        if 10 > p > 0:
            p+=round(p*(-t1),2)
            descuento_aplicado.append(p)
        elif 40 == p >= 10:
            p+=round(p*(-t2),2)
            descuento_aplicado.append(p)
        else:
            p+=round(p*(-t3),2)
            descuento_aplicado.append(p)
    return descuento_aplicado
```

```python
import random
lista_precios_random=[pr for pr in range(3,random.randint(1,66),4)]
print(lista_precios_random)
descuento(lista_precios_random)
```

Encontrar el elemento duplicado en una lista de números.
Descripción: Crea una función llamada encontrar_duplicado que tome una lista de números enteros y devuelva el elemento que está duplicado en la lista.

```python
def encontrar_duplicado(lista):
    frecuencia={}
    duplicados=[]
        
    for n in lista:
        if n not in frecuencia:
            frecuencia[n]=1
        else:
            frecuencia[n]+=1

    for n,v in frecuencia.items():
        if v > 1:
            duplicados.append(n)    
    print(f"Los números repetidos en la lista son: {duplicados}.")
    #return duplicados

```

Calcular el promedio de una lista de números.
Descripción: Crea una función llamada calcular_promedio que tome una lista de números como parámetro y devuelva el promedio de todos sus elementos.

```python
def calcular_promedio(lista):
    a=0
    b=len(lista)
    if b !=0:
        for n in lista:
            a+=n
        print(f"El promedio es '{a/b}'.\nTu lista de números ingresada es: {lista}.")
    else:
        print("Tu lista esta vacía.")

```

```python
lista=[n for n in range(16)]
calcular_promedio(lista)
```

Pista: solo necesitas saber si ambas palabras tienen las mismas letras, sin importar el orden. Puedes usar la función sorted() para ordenar las letras de cada palabra y compararlas.

Validar si una cadena de texto es un isograma.
Descripción: Crea una función llamada es_isograma que tome una cadena de texto como parámetro y devuelva True si es un isograma. En caso contrario, devuelve False.
¿Qué es un isograma? Una palabra que no contiene letras repetidas. Pista: Un set o conjunto no puede tener elementos repetidos.

```python
def es_isograma(texto):
    unicos=set(texto)
    if len(texto) == len(unicos):
        return True
    else:
        return False
```

Crear un diccionario con información de una persona.
Descripción: Crea una función llamada crear_persona que tome como argumentos clave-valor los datos de una persona, como nombre, edad, ciudad, etc., y devuelva un diccionario con toda la información ingresada.

```python
def crear_persona(*args):
    diccionario_persona={}
    if len(args) % 2 == 0: 
        for i in range(0,len(args),2):
            c=args[i]
            v=args[i+1]
            diccionario_persona[c]=v
        return diccionario_persona

    else:
        print("El número de argumentos es impar, intentalo de nuevo ejecutando el programa.")
```

Determinar si una cadena es un palíndromo.
Descripción: Crea una función llamada es_palindromo que tome una cadena como parámetro y devuelva True si es un palíndromo, o False si no lo es.

```python
def es_palindromo(palabra1,palabra2):
   
    palabra_r=palabra2[::-1].lower()
    
    if palabra1.lower() == palabra_r:
        return True
    else:
        return False
```

```python
es_palindromo("amor","ROMA")
```
