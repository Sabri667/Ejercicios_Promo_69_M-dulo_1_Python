# Módulo 1 Tuplas: Nivel Inicial

```python
#Definir tuplas
#Crea una tupla persona1 con los siguientes datos: 61, 1.54, "pelo marrón". (Serán del tipo int, float y str.)
persona1 = (61, 1.54, "pelo marrón")
#Crea una tupla para la persona2 de 68, 1.65 y "pelo rubio".
persona2 = (68, 1.65, "pelo rubio")
#Crea una lista talla que recompila 'M', 42, y 39.
talla = ['M', 42, 39]
#Convertir y juntar 4. Convierte la lista talla a una tupla. 5. Une a la tupla de persona1 los datos en talla, para tener una única tupla.
talla_tupla = tuple(talla)
persona1 = persona1 + talla_tupla
print(persona1)
#Indexar tuplas
#Saca de persona2 el contenido del tercer elemento y llámalo color.
persona2_lista = list(persona2)
print(persona2_lista)
color = persona2_lista.pop(2)
print(color)
#color= persona2. (2,"pelo rubio")
#print(color)
print(persona2_lista)
#¿Cuántos elementos hay ahora en persona1? (Es decir, ¿cuál es su longitud?)
print(len(persona1))
print(len(persona2))
print(len(persona2_lista))
```

```python
#Desempaquetar 8. Crea las variables peso, altura, color_pelo y llénalas con el contenido de persona1. Imprime color_pelo para comprobar el resultado.
peso=persona1[0]
altura=persona1[1]
color_pelo=persona1[2]
print(color_pelo)
#Zipear 9. Crea el zip personas con persona1 y persona2.
personas=list(zip(persona1,persona2))
print(personas)
#¿Cuántas "elementos" tiene personas? ¿Por qué?
len(personas)
#Crea una tupla llamada etiquetas con las variables 'peso', 'altura', y 'color_pelo' que hemos desempaquetado antes.
etiquetas = ('peso', 'altura', 'color_pelo')
#Crea un zip llamado "todo2 que contenta etiquetas, persona1, y persona2.
todo2 = list(zip(etiquetas, persona1, persona2))
print(todo2)
#Manipular zips 13. Ordena las propiedades de la variable personas alfabéticamente. ¿Puedes? ¿Por qué?
sorted_personas = sorted(todo2)
print(sorted_personas)
""" no es posible ordenar la variable personas alfabéticamente porque contiene elementos de diferentes tipos de datos (int, float, str), lo que impide la comparación directa entre ellos. """
```

# Modulo 1 Tuplas: Nivel Intermedio

```python
#Acceder al primer elemento de una tupla
#Descripción: Crea una tupla de números y obtén el primer elemento.
tupla=4,9,12,10
print(tupla[0])
#Acceder al último elemento de una tupla
#Descripción: Ahora obtén el último elemento.
print(tupla[-1])

#Obtener la longitud de una tupla
#Descripción: Ahora obtén la cantidad de elementos que contiene.
print(len(tupla))
#Concatenar dos tuplas
#Descripción: Crea dos tuplas nuevas y luego crea una nueva tupla que contenga todos los elementos de ambas tuplas.
tupla1=8,7,44,2
tupla2=100,2,6.7,112
tupla3=tupla1+tupla2
print(tupla1)
print(tupla2)
print(tupla3)
#Encontrar el índice de un elemento en una tupla
#Descripción: De tu primera tupla, obtén el índice del valor que decidas.
print(tupla1.index(44))
#Contar ocurrencias de un elemento en una tupla
#Descripción: Ahora obtén cuántas veces aparece ese valor en tu tupla.
print(tupla1.count(44))
#Obtener una subtuple de una tupla
#Descripción: Crea una nueva tupla que contenga los elementos desde el índice 1 hasta el índice 3 (sin incluirlo).
tupla_slicing=tupla1[1:3]
print(tupla_slicing)
#Invertir una tupla
#Descripción: Invierte el orden de tu tupla.
tupla1_invertida=tupla1[::-1]
print(tupla1_invertida)
```
