Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias de cada letra en la cadena. Los espacios no deben ser considerados.

```python
            
def frecuencias_letra(texto):
    texto_cadena = texto.strip().replace(" ", "").replace(",","").lower()
    frecuencias = {}

    for letra in texto_cadena:
        if letra in frecuencias:
            frecuencias[letra] += 1
        else:
            frecuencias[letra] = 1

    return frecuencias
```

```python
frecuencias_letra("hola, mundo")
```

Escribe una función que reciba una cadena de texto como parámetro y devuelva una lista con todas las palabras que tienen más de 5 letras.

```python

def palabras_mas_5_letras(texto):
    palabras = texto.split()
    lista_palabras = []
 
    for i in palabras:
        if len(i) > 5:
            lista_palabras.append(i)
        else:
            continue
    return lista_palabras

```

```python
palabras_mas_5_letras("Bienvenidos al nuevo curso de Data Analyst")

```


Crea una función que tome una lista de palabras como entrada y devuelva una lista de palabras que contengan más de una vocal.

```python
def palabras_mas_1_vocal(lista):
    
    lista_palabras=[]
    for palabra in lista:
        cuenta=0
        for letra in palabra:
            if letra in "aeiouAEIOU":
                cuenta+=1
        if cuenta > 1:
            lista_palabras.append(palabra)
    return lista_palabras

```

```python
lista=["sol","luna","mar","paseo","galaxia", "ventana"]
palabras_mas_1_vocal(lista)
```

Crea una función que tome una lista de números y calcule la suma de todos los números pares en la lista.

```python
def suma_pares(lista):
    
    suma_pares=0
    for n in lista:
        if n%2 ==0:
            suma_pares+=n
    return suma_pares


```

```python
lista_numeros=list(range(20,67,3))
print(lista_numeros)
suma_pares(lista_numeros)
```

Escribe una función que tome una lista de números como entrada y calcule el promedio de esos números.

```python
def promedio(lista):

    total = 0
    for n in lista:
        total += n
    try:
        return total / len(lista)
    except ZeroDivisionError:
        return None
```

```python
lista_numeros=list(range(20,35,4))
print(lista_numeros)
promedio(lista_numeros)
```
