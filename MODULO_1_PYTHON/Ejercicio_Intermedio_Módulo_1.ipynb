```python
"""2. Dada la siguiente lista: frutas = ['manzana', 'pera', 'banana', 'uva', 'kiwi'],
realiza lo siguiente:
. Reemplaza el segundo elemento con 'durazno'.
. Agrega 'naranja' al final de la lista.
. Extrae y muestra por pantalla el tercer elemento de la lista. """

frutas = ['manzana','pera', 'banana', 'uva', 'kiwi']
print(f"La lista de frutas incluye: {frutas}")
frutas[1]='durazno'
print(f"Se reemplazó 'pera' por 'durazno': {frutas}")
frutas.extend(['naranja'])
print(f"Se agregó 'naranja' al final de la lista: {frutas}")
print(frutas[2])
```

```python
"""3. Crea una lista llamada numeros que contenga los siguientes valores: 10, 25, 15, 30, 20.
. Agrega el valor 35 al incio de la lista numeros.
. Ordena la lista numeros en orden ascendente.
. Encuentra el índice del número 25."""

numeros=[10,25,15,30,20]
print(f"Se creó la lista numeros = {numeros}")
numeros.insert(0,35)
print(f"Se insertó en la primera posición el número '35': {numeros}")
numeros.sort()
print(f"Se ordenó la lista de forma ascendente: {numeros}")
#numeros.index(25)
print(f"El número '25' se encuentra en la posición {numeros.index(25)} de la lista.")
```

```python
"""4. Crea un diccionario llamado frutas con las siguientes parejas clave-valor:
'manzana': 2
'banana': 3
'uva': 5
'pera': 4
Saca una lista de todas las frutas en el diccionario.
Saca una lista de todas las cantidades en el diccionario.
¿Qué cantidad de 'manzana' tenemos?. Si la clave 'manzana' no está presente, devuelve un mensaje
diciendo "No existe esa clave".
Agrega un nuevo par de clave-valor al diccionario frutas: 'kiwi' con una cantidad de 6.
Cambia la cantidad de 'uva' a 8.
Crea un segundo diccionario llamado frutas_nuevas con las siguientes parejas clave-valor: -
'sandía': 7 - 'mango': 4
Combinar los dos diccionarios frutas_nuevas y frutas en uno nuevos llamado frutas_total.
Elimina el último par de clave-valor del diccionario fruta_total.
Elimina el par de clave-valor correspondiente a la clave 'uva' del diccionario.

Combinar métodos de listas y diccionarios:
Obtén una lista de todas las cantidades en el diccionario y calcula suma total de las cantidades
del diccionario frutas_total.
Utilizando la lista creada en el apartado anterior encuentra la cantidad más grande y más
pequeña."""

frutas={'manzana':2,'banana':2,'uva':5,'pera':4}
print(f"Se creó el diccionario 'frutas': {frutas}")
print(f"Las frutas que incluye son: {list(frutas.keys())}")
print(f"Las cantidades que incluye son: {list(frutas.values())}")
print(f"La cantidad de 'manzana' que tenemos es la lista es {frutas.get('manzana',"No existe esa clave")}.")
frutas['kiwi']=6
print(f"Se agregó 'kiwi' al diccionario: {frutas}")
frutas['uva']=8
print(f"Se cambió las unidades de 'uva' de 5 a 8: {frutas}")
frutas_nuevas={'sandia': 7, 'mango': 4}
print(f"Se creó un nuevo diccionario: 'frutas_nuevas' = {frutas_nuevas}")
frutas_total={**frutas,**frutas_nuevas}
print(f"Se creó un nuevo diccionario que incluye 'frutas' y 'frutas_nuevas', 'frutas_total' = {frutas_total}")
frutas_total.popitem()
print(f"Se eliminó la última 'clave-valor': {frutas_total}")
frutas_total.pop('uva')
print(f"Se eliminó 'uva' del diccionario: {frutas_total}")
lista_cantidades=list(frutas_total.values())
print(f"Se creó una lista con las cantidades del diccionario 'frutas_total': {lista_cantidades}")
print(f"La suma de las cantidades es igual a {sum(lista_cantidades)}.")
print(f"El máximo de unidades en 'lista_cantidades' es: {max(lista_cantidades)}.")
print(f"El mínimo de unidades en 'lista_cantidades' es: {min(lista_cantidades)}.")
```

```python
"""5.Evaluación de Edades:
Escribe un programa que solicite al usuario ingresar su edad.
Si la edad es menor de 18, muestra "Eres menor de edad".
Si la edad está entre 18 y 65 (incluyendo ambos extremos), muestra "Eres adulto".
Si la edad es mayor que 65, muestra "Eres un adulto mayor"""

while True:

    try:

        edad=int(input("Hola,ingresa tu edad por favor: ").strip())
        if edad < 0:
            print("La edad ingresada no es válida. No puede ser menor a 0.")
            continue
        elif edad < 18:
            print(f"Eres menor de edad.")
            break
        elif edad < 65:
            print(f"Eres adulto.")
            break
        elif edad > 65:
            print("Eres un adulto mayor.")
            break
    except ValueError:
        print("Parece que no has ingresado un número válido! Si quieres volver a intentarlo ejecuta de nuevo el programa.")
        break

```

```python
"""6. Determinación de Día de la Semana
Solicita al usuario que ingrese un número del 1 al 7.
Utilizando una serie de if-elif-else, muestra el día de la semana correspondiente al
número ingresado. Por ejemplo, si ingresa 1, muestra "Lunes"."""

dias_semana=['Lunes','Martes','Miércoles','Jueves','Viernes', 'Sábado','Domingo']

while True:
    try:
        numero_dia=input("Indica un número del 1 al 7: ".strip())
        if numero_dia in '1234567':
            numero_dia=int(numero_dia)
            numero_dia-=1
            print(dias_semana[numero_dia])
            break
        else:
            print(f"Por favor, ingresa un número válido del 1 al 7.")
            continue
    except ValueError:
        print("Parece que los caracteres ingresados no son numéricos. Ejecuta este programa nuevamente si quieres volver a intentarlo.")
        break
```

```python
"""7.Juego de Adivinanza
Genera un número aleatorio entre 1 y 100.
Pídele al usuario que adivine el número.
Si el usuario adivina correctamente en el primer intento, muestra "¡Ganaste en el primer
intento!".
Si el usuario no adivina en el primer intento, bríndale pistas (mayor o menor) y permite que
intente nuevamente. Muestra cuántos intentos le llevó al final."""
import random
n=random.randint(1,100)
intentos=1
print(n) #para control
while True:

    try:
        n_adivinando=int(input("Elige un número del 1 al 100 y veremos si has adivinado el número secreto!: "))

        if n_adivinando < 1  or n_adivinando >100:
            print("Debes indicar un número entre 1 y 100.")
            continue
        elif n_adivinando == n and intentos == 1:
            print("¡Ganaste en el primer intento!")
            break
        elif n_adivinando > n:
            intentos+=1
            print("¡Pista! El número secreto es menor.")
            continue
        elif n_adivinando < n:
            intentos +=1
            print("¡Pista! El número secreto es mayor.")
            continue
        else:
            print(f"¡¡Lo has adivinado!! Te tomó {intentos} intentos.")
            break
        
    except ValueError:
        print("Debes ingresar un número del 1 al 100. Si quieres volver a intentarlo ejecuta nuevamente el programa.")
        break




```
