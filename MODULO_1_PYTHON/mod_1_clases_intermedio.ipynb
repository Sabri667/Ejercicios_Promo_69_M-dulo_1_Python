```python
class Empleados:
    """Clase base que representa a un empleado genérico. Contiene atributos básicos y métodos para gestionar información laboral."""

    def __init__(self, nombre, apellidos, edad, posicion, año, vacaciones, herramientas, salario=1500):
        """ Constructor de la clase Empleados. Inicializa los atributos del empleado.     
        Parámetros:
        - nombre: str, nombre del empleado.
        - apellidos: str, apellidos del empleado.
        - edad: int, edad del empleado.
        - posicion: str, cargo o puesto del empleado.
        - año: int, año de ingreso a la empresa.
        - vacaciones: int, días de vacaciones disponibles.
        - herramientas: list, lista de tecnologías o habilidades que maneja.
        - salario: float o int, salario del empleado. Por defecto 1500.
        """
        self.nombre = nombre
        self.apellidos = apellidos
        self.edad = edad
        self.posicion = posicion
        self.año = año
        self.vacaciones = vacaciones
        self.herramientas = herramientas
        self.salario = salario

    def descripcion(self):
        """ Muestra una descripción completa del empleado: nombre completo, puesto, año de ingreso, días de vacaciones y salario formateado. """
        print(f"{self.nombre.title().strip()} {self.apellidos.title().strip()}, es {self.posicion.title().strip()}.\nIngresó en {self.año} y actualmente posee {self.vacaciones} días de vacaciones para usufructuar este año.\nSu salario es de ${self.salario:,.2f}.")

    def conocimientos_tecnicos(self):
        """ Revisa la lista de herramientas del empleado. Si tiene "Python", lo felicita y termina. Si tiene "Excel", sugiere aprender Python. Para otras herramientas, pregunta si le interesaría incorporar alguna nueva.
        Si el empleado no tiene apellidos (o no está en nómina), informa que no está en nómina. """
        if self.apellidos:
            for h in self.herramientas:
                if h.lower().strip() == "python":
                    print(f"¡Ya cuentas con Python en tu saber, muy bien!")
                    break
                elif h.lower().strip() == "excel":
                    print(f"Veo que sabes excel... ya puedes dar el salto e incorporar Python!")
                    continue
                else:
                    print(f"¿Te interesaría incorporar alguna nueva herramienta?")
        else:
            print(f"El empleado {self.apellidos} no se encuentra en la nómina.")

    def calcular_antiguedad(self):
        """ Calcula y muestra los años de antigüedad del empleado. Valida que el año de ingreso sea un número de 4 dígitos, mayor o igual a 2016,
        y no mayor al año actual. Luego calcula la diferencia con el año actual. """
        import re
        from datetime import datetime, date

        match = re.search(r"\d{4}", str(self.año))
        año_actual = datetime.now().year
        if not match:
            print("Año no válido (ingrese número de 4 dígitos).")
            return

        año_ingreso = int(match.group())
        if año_ingreso < 2016:
            print(f"Ingrese un año igual o mayor a 2016.")
            return
        if año_ingreso > año_actual:
            print(f"El año de ingreso no puede ser mayor al actual.")
            return

        antiguedad = date.today().year - año_ingreso
        print(f"El empleado {self.apellidos} tiene {antiguedad} año(s) de antigüedad.")

    def solicitar_aumento_vacaciones(self, dias_solicitados):
        """ Aumenta los días de vacaciones del empleado.         
        Parámetros:
        - dias_solicitados: int, cantidad de días a añadir (debe ser positivo).
        Muestra mensaje de éxito o error según corresponda.
        """
        if dias_solicitados and isinstance(dias_solicitados, int):
            if dias_solicitados > 0:
                self.vacaciones += dias_solicitados
                print(f"Su solicitud fue procesada con éxito. Ahora cuenta con {self.vacaciones} días de vaciones.")
            else:
                print("No se pueden procesar días negativos.")
        else:
            print("Para procesar tu solicitud ingresa una cantidad de días a 0 (cero).")

    def registrar_herramienta(self, nueva_herramienta):
        """ Agrega una nueva herramienta a la lista del empleado si no existe ya.
        
        Parámetros:
        - nueva_herramienta: str, nombre de la herramienta a registrar.
        Si ya existe, informa que está duplicada. Si no, la agrega con formato título.
        """
        for h in self.herramientas:
            if nueva_herramienta.title().strip() in h.title().strip():
                print(f"Lo siento, la herramienta que estas intentado incorporar '{nueva_herramienta}', ya se encuentra registrada.")
                return
        else:
            self.herramientas.append(nueva_herramienta.title().strip())
            print(f"Se ha incorporado '{nueva_herramienta}', a tu lista de herramientas.")

    def cambiar_de_posicion(self, nueva_posicion, nuevo_salario):
        """ Cambia el puesto y el salario del empleado si el nuevo salario es >= 1500.
        
        Parámetros:
        - nueva_posicion: str, nuevo cargo.
        - nuevo_salario: float/int, nuevo salario (debe ser al menos 1500).
        Muestra felicitaciones en caso de ascenso, o error si no cumple condiciones.
        """
        if nueva_posicion and nuevo_salario >= 1500:
            self.posicion = nueva_posicion.title().strip()
            self.salario = nuevo_salario
            print(f"¡Felicitaciones {self.nombre} por el ascenso y lo mejor en tu nuevo puesto como {self.posicion}!")
        else:
            print(f"Debes ingresar una posición y un salario al menos igual al salario base de la empresa.")

    def tomar_vacaciones(self):
        """ Permite al empleado tomar días de vacaciones. Solicita interactivamente la cantidad de días a tomar.
        Valida que sea un número entero positivo. Si los días solicitados superan los disponibles, ofrece tomar los restantes.
        Actualiza el saldo de vacaciones. Maneja errores de entrada no numérica. """
        while True:
            try:
                dias_a_tomar = int(input("Ingresa la cantidad de días a tomarte: "))
                if dias_a_tomar <= 0:
                    print("Debes ingresar una cantidad de días mayor a 0(cero).")
                    continue

                if self.vacaciones == 0:
                    print("No tienes días de vacaciones pendientes. No se puede procesar tu solicitud.")
                    break

                if dias_a_tomar > self.vacaciones:
                    print(f"Solo te quedan {self.vacaciones} días. No puedes tomar {dias_a_tomar}.")
                    respuesta = input(f"¿Quieres tomar los {self.vacaciones} días restantes? (s/n): ")
                    if respuesta.lower().strip() == 's':
                        print(f"Se te autorizaron {self.vacaciones} días de vacaciones. Tienes 0 (cero) días pendientes de vacaciones.")
                        self.vacaciones = 0
                    else:
                        print("Solicitud cancelada.")
                    break
                else:
                    self.vacaciones -= dias_a_tomar
                    print(f"Se han autoriado {dias_a_tomar} día(s). Tienes {self.vacaciones} días pendientes.")
                    break
            except ValueError:
                print("Entrada inválida. Debes ingresar un número entero.")

    def solicitar_aumento_salarial(self):
        """ Solicita al usuario un porcentaje de aumento salarial (máximo 50%). Valida que la entrada sea un número positivo de hasta 2 dígitos (0-50).
        Calcula y actualiza el salario aplicando el porcentaje. Muestra el nuevo salario formateado. Maneja errores de formato y valores fuera de rango. """
        import re

        while True:
            try:
                aumento = input("Ingrese el porcentaje de aumento salarial que quiere se autorice (NN): ")

                if "-" in aumento.strip():
                    print(f"Debes ingresar un número positvo de hasta 2 (dos) dígitos.")
                    continue

                match = re.search(r"\b(\d{1,2}(?:[.,]\d{1,2})?)\s*%?\b", aumento)

                if not match:
                    print("Debe ingresar un número positivo de 2 (dos) dígitos.")
                    continue
                aumento_salarial = int(match.group())
                if aumento_salarial > 50:
                    print(f"No podemos procesar aumentos salariales mayores al 50%. Ingresa un número positvo de hasta 2 (dos) dígitos menor al 50%.")
                    continue
                else:
                    self.salario = self.salario * (1 + (aumento_salarial / 100))
                    print(f"Hemos procesado tu solicitud de aumento salarial del {aumento_salarial} %.\nTu nuevo salario es de $ {self.salario:,.2f}.")
                    break
            except ValueError:
                print("Debe ingresar un número positivo de máximo 2 (dos) dígitos, sin './,'(punto/coma).")
                break

    def realizar_evaluacion(self):
        """ Realiza una autoevaluación del desempeño. Solicita al usuario una calificación entre: 'excelente', 'bueno', 'regular', 'malo'.
        Muestra un mensaje según la calificación. Si la entrada no es válida, vuelve a preguntar. Maneja interrupción por teclado (Ctrl+C) saliendo del método. """
        try:
            while True:
                calificacion = input(f"{self.nombre} ingresa tu calificación de desempeño '('excelente'/'bueno'/'regular'/'malo'): ")
                if not calificacion:
                    continue
                if calificacion.lower().strip() == "excelente":
                    print("¡Felicitaciones! Tu calificación de tu desempeño es excelente. Eres un empleado destacado.")
                    return
                elif calificacion.lower().strip() == "bueno":
                    print("La calificación de tu desempeño es buena. ¡Sigue así!")
                    return
                elif calificacion.lower().strip() == "regular":
                    print("La calificación de tu desempeño es regular. Es momento de tomar medidas para mejorar.")
                    return
                elif calificacion.lower().strip() == "malo":
                    print("La calificación de tu desempeño es mala. Es hora de comprometerse y tomar medidas para mejorar.")
                    return
                else:
                    print("Debes calificar tu desempeño con alguna de las siguientes opciones: 'excelente'/'bueno'/'regular'/'malo'.")
                    continue
        except KeyboardInterrupt:
            print("\nOperación cancelada por el usuario.")
            return
```

```python
class Gerente(Empleados):

    """Clase Gerente que hereda de Empleados."""

    def __init__(self, nombre, apellidos, edad, posicion, año, vacaciones, herramientas, salario=None, departamento=None, empleados_a_cargo=None):
        # No se elimina atributo, pero se asigna valor "Gerente" siempre
        posicion = "Gerente"   
        # Validación de salario
        if salario is None or salario < 2200:
            salario = 2200
            print("El salario no puede ser menor a $2.200. Se asignará el valor base.")
        # Llamada al constructor padre con el salario ya validado
        super().__init__(nombre, apellidos, edad, posicion, año, vacaciones, herramientas, salario)
        self.departamento = departamento
        # empleados_a_cargo es una lista vacía si es None
        self.empleados_a_cargo = empleados_a_cargo if empleados_a_cargo is not None else []

    def asignar_tareas(self, tarea, empleado):
        """Devuelve un mensaje indicando que se asignó la tarea al empleado, siempre que el empleado esté a cargo."""
        
        empleado = empleado.title().strip()
        if empleado in self.empleados_a_cargo:
            return f"El gerente ha asignado la tarea '{tarea}' a '{empleado}'."
        else:
            return f"No se puede asignar la tarea. '{empleado}' no está en la lista de empleados a cargo."

    def revisar_desempeno_empleado(self, empleado):
        """Devuelve un mensaje indicando que se está revisando el desempeño, si el empleado está a cargo."""

        empleado = empleado.title().strip()
        if empleado in self.empleados_a_cargo:
            return f"El gerente está revisando el desempeño de '{empleado}'."
        else:
            return f"'{empleado}' no se encuentra en la lista de empleados a cargo."
        

    def gestionar_empleados(self, accion, empleado):
        """Permite contratar o despedir empleados. accion: 'contratar' o 'despedir', empleado: nombre del empleado (str)
        Devuelve un mensaje según el resultado."""

        empleado = empleado.title().strip()
        accion = accion.lower().strip()

        if accion not in ['contratar', 'despedir']:
            return "Acción no válida. Debe ser 'contratar' o 'despedir'."

        if accion == 'contratar':
            if empleado in self.empleados_a_cargo:
                return f"'{empleado}' ya tiene un contrato vigente en el departamento '{self.departamento}'."
            else:
                self.empleados_a_cargo.append(empleado)   
                return f"Se ha contratado a '{empleado}' en el departamento '{self.departamento}'."

        elif accion == 'despedir':
            if empleado in self.empleados_a_cargo:
                self.empleados_a_cargo.remove(empleado)
                return f"Se ha despedido a '{empleado}' del departamento '{self.departamento}'."
            else:
                return f"'{empleado}' no tiene contrato vigente en el departamento '{self.departamento}'."
    def ver_empleados_a_cargo(self):
        """Devuelve una cadena con la lista de empleados a cargo."""
        if not self.empleados_a_cargo:
            return "No tiene empleados a cargo."
        return f"Empleados a cargo: {', '.join(self.empleados_a_cargo)}"

  



```
