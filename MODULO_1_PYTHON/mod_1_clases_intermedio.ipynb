```python
class Empleados:
    
    def __init__(self, nombre, apellidos, edad, posicion, año , vacaciones, herramientas,salario=1500.):
        self.nombre=nombre
        self.apellidos=apellidos
        self.edad=edad
        self.posicion=posicion
        self.año=año
        self.vacaciones=vacaciones
        self.herramientas=herramientas
        self.salario=salario

    def descripcion(self):
        print(f"{self.nombre.title().strip()} {self.apellidos.title().strip()}, es {self.posicion.title().strip()}.\nIngresó en {self.año} y actualmente posee {self.vacaciones} días de vacaciones para usufructuar este año.\nSu salario es de ${self.salario:,.2f}.")

    def conocimientos_tecnicos(self):
        if self.apellidos:
            for h in self.herramientas:
                if h.lower().strip() == "python":
                    print(f"¡Ya cuentas con Python en tu saber, muy bien!")
                    break
                elif h.lower().strip() == "excel":
                    print(f"Veo que sabes excel... ya puedes dar el salto e incorporar Python!")
                    continue
                else:
                    print(f"¿Te interesaría incorporar alguna nuev herramienta?")
        else:
            print(f"El empleado {self.apellidos} no se encuentra en la nómina.")

    def calcular_antiguedad(self):  #solo años
        import re
        from datetime import datetime, date
    
        match = re.search(r"\d{4}", str(self.año))
        año_actual = datetime.now().year
        if not match:
            print("Año no válido (ingrese número de 4 dígitos).")
            return
        
        año_ingreso = int(match.group())
        if año_ingreso < 2016:
            print(f"Ingrese un año igual o mayor a 2016.")
            return
        if año_ingreso > año_actual:
            print(f"El año de ingreso no puede ser mayor al actual.")
            return
        
        antiguedad = date.today().year - año_ingreso
        print(f"El empleado {self.apellidos} tiene {antiguedad} año(s) de antigüedad.")

    def solicitar_aumento_vacaciones(self,dias_solicitados):
        
        if dias_solicitados and isinstance(dias_solicitados, int):
            if dias_solicitados > 0:
                self.vacaciones+=dias_solicitados
                print(f"Su solicitud fue procesada con éxito. Ahora cuenta con {self.vacaciones} días de vaciones.")
            else:
                print("No se pueden procesar días negativos.")
        else:
            print("Para procesar tu solicitud ingresa una cantidad de días a 0 (cero).")

    def registrar_herramienta(self, nueva_herramienta):
        for h in self.herramientas:
            if nueva_herramienta.title().strip() in h.title().strip():
                print(f"Lo siento, la herramienta que estas intentado incorporar '{nueva_herramienta}', ya se encuentra registrada.")
                return
        else:
            self.herramientas.append(nueva_herramienta.title().strip())
            print(f"Se ha incorporado '{nueva_herramienta}', a tu lista de herramientas.")
    

    def cambiar_de_posicion(self, nueva_posicion, nuevo_salario):
        if nueva_posicion and nuevo_salario >= 1500:    #salario base modelo para límite inferior.
            self.posicion=nueva_posicion.title().strip()
            self.salario=nuevo_salario
            print(f"¡Felicitaciones {self.nombre} por el ascenso y lo mejor en tu nuevo puesto como {self.posicion}!")
        else:
            print(f"Debes ingresar una posición y un salario al menos igual al salario base de la empresa.")
    
    def tomar_vacaciones(self):
        while True:
            try:
                dias_a_tomar = int(input("Ingresa la cantidad de días a tomarte: "))
                if dias_a_tomar <= 0:
                    print("Debes ingresar una cantidad de días mayor a 0(cero).")
                    continue
                
                if self.vacaciones == 0:
                    print("No tienes días de vacaciones pendientes. No se puede procesar tu solicitud.")
                    break 

                if dias_a_tomar > self.vacaciones:
                
                    print(f"Solo te quedan {self.vacaciones} días. No puedes tomar {dias_a_tomar}.")
                    respuesta = input(f"¿Quieres tomar los {self.vacaciones} días restantes? (s/n): ")
                
                    if respuesta.lower().strip() == 's':
                        print(f"Se te autorizaron {self.vacaciones} días de vacaciones. Tienes 0 (cero) días pendientes de vacaciones.")
                        self.vacaciones = 0
                    else:
                        print("Solicitud cancelada.")
                    break  

                else:
                    self.vacaciones -= dias_a_tomar
                    print(f"Se han autoriado {dias_a_tomar} día(s). Tienes {self.vacaciones} días pendientes.")
                    break
                
            except ValueError:
                print("Entrada inválida. Debes ingresar un número entero.")
                

    def solicitar_aumento_salarial(self):
         import re

         while True:
             try:
                 aumento=input("Ingrese el porcentaje de aumento salarial que quiere se autorice (NN): ")
                                  
                 if "-" in aumento.strip():
                    print(f"Debes ingresar un número positvo de hasta 2 (dos) dígitos.")
                    continue
                 
                 match=re.search(r"\b(\d{1,2}(?:[.,]\d{1,2})?)\s*%?\b",aumento)

                 if not match:
                     print("Debe ingresar un número positivo de 2 (dos) dígitos.")   #test: 3 digitos, numeros negativos, limite superior, . , %
                     continue
                 aumento_salarial=int(match.group())
                 if aumento_salarial > 50:   #límite superior
                    print(f"No podemos procesar aumentos salariales mayores al 50%. Ingresa un número positvo de hasta 2 (dos) dígitos menor al 50%.")
                    continue
                 else:
                    self.salario=self.salario*(1+(aumento_salarial/100))
                    print(f"Hemos procesado tu solicitud de aumento salarial del {aumento_salarial} %.\nTu nuevo salario es de $ {self.salario:,.2f}.")
                    break
                     
             except ValueError:
                 print("Debe ingresar un número positivo de máximo 2 (dos) dígitos, sin './,'(punto/coma).")
                 break
      
    def realizar_evaluacion(self):
        import signal

        try:
            while True:
                calificacion=input(f"{self.nombre} ingresa tu calificación de desempeño '('excelente'/'bueno'/'regular'/'malo'): ")
                if not calificacion:
                    continue
                if calificacion.lower().strip() == "excelente":
                    print("¡Felicitaciones! Tu calificación de tu desempeño es excelente. Eres un empleado destacado.")
                    return
                elif calificacion.lower().strip() == "bueno":
                    print("La calificación de tu desempeño es buena. ¡Sigue así!")
                    return
                elif calificacion.lower().strip() == "regular":
                    print("La calificación de tu desempeño es regular. Es momento de tomar medidas para mejorar.")
                    return
                elif calificacion.lower().strip() == "malo":
                    print("La calificación de tu desempeño es mala. Es hora de comprometerse y tomar medidas para mejorar.")
                    return
                else:
                    print("Debes calificar tu desempeño con alguna de las siguientes opciones: 'excelente'/'bueno'/'regular'/'malo'.")
                    continue
        except KeyboardInterrupt:
            print("\nOperación cancelada por el usuario.")
            return
```
