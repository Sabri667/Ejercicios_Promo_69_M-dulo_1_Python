# Módulo 1 Sets: Nivel Inicial

```python
#Define la lista

temperaturas_diciembre = [10, 10, 13, 14, 12, 15,
                          12, 11, 13, 17, 18, 15,
                          15, 17, 17, 13, 13, 11,
                          12, 10, 12, 14, 12, 12,
                          11, 12, 16, 16, 14, 17, 18]
#Conviértala en un set llamado diciembre.
diciembre=set(temperaturas_diciembre)
print(diciembre)
#Piensa en 6 colores y almacenalos en un set llamado colores.
colores={"violeta","azul","amarillo","verde","negro","blanco"}
#Añade 'lila' a colores.
colores.add("lila")
print(colores)
#Añade 'azul', 'rojo', y 'granate' a colores usando .update().
colores.update(["azul","rojo","granate"])
print(colores)
#Tu set de colores, ¿tiene 'verde'?
"verde" in colores

```

```python
#¿Y tiene 'ROJO'?
"ROJO" in colores
#Define el set enero para que sea igual a diciembre.
enero=set(diciembre)
print(enero)
#Quita las temperaturas 17, y 18 de enero. 
enero.discard(17)
enero.discard(18)
print(enero)
#¿Cuántas temperaturas distintas había en diciembre?
print(len(diciembre))
#¿Y cuántas en enero?
print(len(enero))
#Identifica los elementos que hay en diciembre que no están en enero.
print(diciembre - enero)
#Identifica los elementos que hay en enero que no están en diciembre.
print(enero - diciembre)
#Identifica los elementos que tienen en común diciembre y enero.
print(diciembre & enero)
#Identifica la diferencia simétrica de colores y {'marrón','amarillo','cyan'}. ¿Qué significa este resultado?
colores2={"marrón","amarillo","cyan"}
print(colores ^ colores2)
#¿Es {'rojo','azul'} un superset de colores?
print({"rojo","azul"}.issuperset(colores))
print({"rojo","azul"} >= colores)

#¿Y un subset?
print({"rojo","azul"}.issubset(colores))
print({"rojo","azul"} <= colores)
```

# Módulo 1 Sets: Nivel Intermedio

```python
#Crear un conjunto vacío
#Descripción: Crea un conjunto vacío sin ningún elemento.
set_vacio=set()
print(set_vacio)

#Crear un conjunto con elementos
#Descripción: Crea un conjunto con elementos predefinidos.
set_días={"lunes","martes","miércoles","jueves","viernes","sábado","domingo"}
print(set_días)
#Agregar elementos a un conjunto
#Descripción: Usando el set creado en el ejercicio 2, agrega nuevos elementos. Debereis agregar por separado:
#Un elemento solo
set_días.add("festivo")
print(set_días)
#Un par de elementos a la vez.
set_días.update(["laborable","vacaciones"])
print(set_días)
#Eliminar un elemento de un conjunto
set_días.remove("vacaciones")
print(set_días)
#Descripción: Usando el set creado en el ejercicio 2, elimina un elemento especificado por la usuaria (input).
remover_de_set_dias=input("Introduce el día para eliminarlo del conjunto: ")
set_días.remove(remover_de_set_dias)
print(set_días)
```

```python
#Verificar la pertenencia de un elemento en un conjunto
#Descripción: Usando el set creado en el ejercicio 2, verifica si un elemento específico está presente en él.
"jueves" in set_días
#Realizar una unión de conjuntos
#Descripción: Dados dos conjuntos, realiza una unión de ellos.
set_sucursales={"Madrid","Barcelona","Valencia"}
set_nuevas_sucursales={"Sevilla","Zaragoza","Madrid"}
set_sucursales2026=set_sucursales | set_nuevas_sucursales
print(set_sucursales2026)
#Realizar una intersección de conjuntos
#Descripción: Usandolos dos conjuntos de datos definidos en el ejercicio anterior, realiza una intersección de ellos.
print(set_sucursales & set_nuevas_sucursales)
#Realizar una diferencia de conjuntos
#Descripción: Usando los dos conjuntos de datos definidos en el ejercicio anterior, realiza una diferencia entre ellos.
print(set_sucursales.symmetric_difference(set_nuevas_sucursales))
#Verificar si un conjunto es subconjunto de otro
#Descripción: Usando los dos subconjuntos anteriores, verifica si uno es subconjunto del otro.
print(set_sucursales.issubset(set_nuevas_sucursales))
print(set_nuevas_sucursales.issubset(set_sucursales))
print(set_sucursales <= set_nuevas_sucursales)
print(set_nuevas_sucursales <= set_sucursales)
```
