```python
#Calcular la suma de los primeros 10 números naturales utilizando un bucle while
#Descripción: Utiliza un bucle while para calcular la suma de los primeros 10 números naturales (1 + 2 + 3 + ... + 10).

contador=1
total_suma=0
while contador <= 10:
    total_suma += contador
    contador += 1
print(f"La suma de los primeros 10 números naturales es: {total_suma}")
#Juego de adivinanza utilizando un bucle while
#Descripción: Crea un juego de adivinanza en el que el jugador tiene que adivinar un número secreto (tendrá que ser entre el 0 y el 10). Utiliza un bucle while para permitir múltiples intentos hasta que el jugador adivine correctamente.
import random
numero_secreto=random.randint(0,10)
intentos_válidos=0
while True:
    numero_adivinado=int(input("Adivina el número secreto (entre 0 y 10): "))
    if numero_adivinado < 0 or numero_adivinado > 10:
        print("Número no válido. Por favor, introduce un número entre 0 y 10.")
        continue
    intentos_válidos += 1
    if numero_adivinado < numero_secreto:
        print("El número secreto es mayor.")
        continue
    elif numero_adivinado > numero_secreto:
        print("El número secreto es menor.")
        continue
    else:
        print(f"¡Felicidades! Adivinaste el número en {intentos_válidos} intento/s.")
        break
```
