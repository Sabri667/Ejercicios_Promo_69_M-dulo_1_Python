
Imprimir los números primos del 1 al 50 utilizando un bucle for
Descripción: Utiliza un bucle for para imprimir los números primos del 1 al 50 en orden ascendente.
NOTA: Un número primo es un número natural mayor que 1 que solo se solo da 0 al dividirlo por 1 y por sí mismo. 

```python
lista_1_50=list(range(1,51))
n_primos=[]

for i in lista_1_50:

    if i>1:
        for divisor in range(2,int(i**0.5)+1):
            if (i%divisor) == 0:
                break
        else:
            n_primos.append(i)
print(n_primos)

```

Calcular el factorial de un número utilizando un bucle for
Descripción: Dado un número entero positivo, calcula su factorial utilizando un bucle for. El factorial de un número se calcula multiplicando todos los números enteros positivos desde 1 hasta ese número. Por ejemplo, el factorial de 5 es 5 * 4 * 3 * 2 * 1 = 120.

```python
import random
a=random.randint(1,100)
print(a)
f=1
for i in range(a+1):
    if i == 1:
        f=1
    else:
        f=f*i
print(f"El factorial de {a} es {f}.")



        
    
```

Juego de adivinar palabras utilizando un bucle for
Descripción: Crea un juego en el que el jugador debe adivinar una palabra secreta. El jugador tiene 5 intentos, si no lo consigue antes, el juego se acaba. Utiliza un bucle for para controlar los intentos del jugador.

```python
import random

lista_palabras = ["sol", "planeta", "ciudad", "gato", "matemáticas", "pizza"]
palabra_secreta = random.choice(lista_palabras)

intentos_maximos = 5

print("Adivina la palabra secreta.")

for intento in range(1, intentos_maximos + 1):
    palabra = input(f"Intento {intento} de {intentos_maximos}: ").strip().lower()
    
    if palabra == palabra_secreta:
        print(f"¡Felicitaciones! Adivinaste en {intento} intento(s).")
        break
    else:
        restantes = intentos_maximos - intento
        if restantes > 0:
            print(f"Incorrecto. Te quedan {restantes} intento(s).")
        else:
            print(f"Agotaste tus {intentos_maximos} intentos. La palabra secreta era '{palabra_secreta}'.")
```

## Ejercicios _list comprehension_


Generar una lista de los números pares del 1 al 20.
Descripción: Utiliza list comprehension para generar una lista que contenga todos los números pares en el rango del 1 al 20.

```python
numeros_pares_20= [n for n in range(1,21) if n % 2 == 0]
print(numeros_pares_20)
```

Obtener una lista de las vocales en una cadena de texto.
Descripción: Utiliza list comprehension para crear una lista que contenga todas las vocales presentes en una cadena de texto (sin repetición).

```python
texto=input("Ingresa una frase: ").lower()
vistas=[]

vocales_texto=[v for v in texto if v in "aeiou" and v not in vistas and not vistas.append(v)]
print(vocales_texto)
```

Filtrar los números positivos de una lista.
Descripción: Crea una lista con números positivos y negativos. Utiliza list comprehension quedarte solo con los positivos.

```python
numeros=list(range(-10,11))
print(numeros)
numeros_positivos=[n for n in numeros if n >= 0]
print(numeros_positivos)
```

#Ejercicios try... except

Realizar una búsqueda en una lista y capturar una excepción en caso de valor no encontrado.
Descripción: Solicita a la usuaria (input) una lista de nombres y un nombre para buscar en la lista. Intenta encontrar el nombre en la lista y utiliza las sentencias try y except para capturar la excepción en caso de que el nombre no sea encontrado.

```python
while True:
    try:
        texto=input("Indica una lista de nombres separada por comas (','): ").replace(" ","").strip().lower()
        lista = [n for n in texto.split(",") if n]
        if not lista:
            continue

        while True:
            nombre=input("Indica el nombre a buscar: ").lower().strip()
            if nombre not in lista:
                print(f"El nombre que igresaste '{nombre}', no se encuentra en la lista. Este es el listado de nombres: {lista}")
                continue
            else:
                print(f"El nombre que buscas, '{nombre}', se encuentra en la posición número {lista.index(nombre)+1}.")
                break
        break

    except:
        break            
```

```python
while True:
    try:
        texto = input("Indica una lista de nombres separada por comas (','): ").replace(" ", "").strip().lower()
        lista = [n for n in texto.split(",") if n]
        if not lista:
            continue

        while True:
            nombre = input("Indica el nombre a buscar: ").lower().strip()
            try:
                posicion = lista.index(nombre)
                print(f"El nombre que buscas, '{nombre}', se encuentra en la posición número {posicion + 1}.")
                break
            except ValueError:
                print(f"El nombre '{nombre}' no se encuentra en la lista: {lista}")
                continue
        break

    except:
        break         
```
