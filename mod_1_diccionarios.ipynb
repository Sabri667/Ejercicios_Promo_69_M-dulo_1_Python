# Módulo 1 Diccionarios: Nivel Inicial

```python
#Crea el diccionario festivos con las fechas como claves y los nombres como valores: 01/01: Año Nuevo, 06/01: Epifanía del Señor, 15/04: Viernes Santo, 15/08: Asunción de la Virgen, 12/10: Fiesta Nacional de España, 01/11: Todos los Santos, 06/12: Día de la Constitución Española, y 08/12: Inmaculada Concepción. Todos los valores son de tipo string.
festivos=dict({"01/01":"Año Nuevo", "06/01":"Epifanía del Señor", "15/04": "Viernes Santo", "15/08": "Asunción de la Virgen", "12/10": "Fiesta Nacional de España", "01/11": "Todos los Santos", "06/12": "Día de la Constitución Española", "08/12": "Inmaculada Concepción"})

#Crea el diccionario monarcas con los países como claves y los nombres como valores: Bélgica: Philippe, Dinamarca: Margrethe II, Holanda: Beatrix, Noruega: Harald V, España: Felipe VI, Reino Unido: Carlos III.
monarcas=dict({"Bélgica":"Philippe", "Dinamarca":"Margrethe II", "Holanda":"Beatrix", "Noruega":"Harald V", "España":"Felipe VI", "Reino Unido":"Carlos III"})
#¿Qué festivo hay el 15/04?
print(festivos["15/04"])
#La reina 'Beatrix' abdicó, ahora es el rey 'Willem-Alexander'. Actualiza el dict.
"Beatrix" in monarcas.values()

def buscar_todas_claves(monarcas, valor_buscado):
    claves = []
    for clave, valor in monarcas.items():
        if valor == valor_buscado:
            claves.append(clave)
    return claves

print(buscar_todas_claves(monarcas, "Beatrix"))  
monarcas["Holanda"] = "Willem-Alexander"
#El rey de Luxemburgo se llama Henri. Añadelo al dict usando .update().
monarcas.update({"Luxemburgo":"Henri"})
#¿Cuántos países están definidos en monarcas?
print(len(monarcas))
#¿Es 'Monaco' uno de ellos? Compruébalo de una manera en que obtengas un True o False.
"Monaco" in monarcas
```

```python
#¿Es verdad que nos olvidamos de poner 'Suecia'? Compruébalo de una manera en que obtengas un True o False.
from posixpath import sep

"Suecia" in monarcas
#Saca una lista de todos los países en el dict monarcas.
print(*monarcas.keys(), sep='\n') #uno por renglón
#Saca una lista de los nombres de los festivos.
print(list(festivos.values()))
#Imprime el diccionario de festivos.
print(festivos)
#Imprime una lista de tuplas de todos los elementos que hay en festivos.
print(list(festivos.items()))
#Copia monarcas a monarcas2010.
monarcas2010=monarcas.copy()
print(monarcas2010)
#Quita España de monarcas2010.
monarcas2010.pop("España")
print(monarcas2010)
#Quita el último elemento de monarcas2010.
monarcas2010.popitem()
print(monarcas2010)
#Vacia el dict monarcas2010.
monarcas2010.clear()
print(monarcas2010)
#¿Qué festivo hay el 06/01?
print(festivos["06/01"])
#Obten el festivo que haya el 08/03, y si no existen, en su defecto devuélve 'Día Internacional de la Mujer'.
print(festivos.get("08/03", "Día Internacional de la Mujer"))
#Obten el festivo que haya el 31/10, y si no existe crea un par de clave-valor nuevo con el valor de 'Halloween'. Pista: Investiga cómo funciona .setdefault()
festivos.setdefault("31/10", "Halloween")
print(festivos)
```

# Módulo 1 Diccionarios: Nivel Intermedio

```python
#Crear un diccionario vacío
#Descripción: Crea un diccionario vacío sin ningún par clave-valor.
diccionario_vacio = dict()
print(diccionario_vacio)
#Crear un diccionario con elementos
#Descripción: Crea un diccionario con pares clave-valor predefinidos. Definir el diccionario con las claves y valores que querais. NOTA El diccionario debe tener al menos 5 pares de clave:valor.
cursos=dict({"Python": ["Jesica", "Carolina", "Andrea"], "JavaScript": ["Sonia"], "HTML": ["Sonia"], "CSS": ["Ana", "Jesica"], "Power BI": [],"Tableau": ["Carolina","Sonia","Emanuel","Javier"]})
#Acceder al valor de una clave en un diccionario
cursos["Power BI"]
cursos["Python"]
#Descripción: Usando el diccionario creado en el paso anterior, obtén el valor correspondiente a una clave específica que indique la usuaria (debereis usar la función input() para preguntar).
consultar_alumnos = input("Indica un curso para saber los alumnos inscriptos: ")
print(cursos[consultar_alumnos])
#Agregar un nuevo par clave-valor a un diccionario
#Descripción: Usando el diccionario creado en el ejercicio 2, agrega un nuevo par clave-valor.
cursos["SQL"] = ["Sandra", "José", "Marta", "Pedro"]
print(cursos)
#Actualizar el valor de una clave en un diccionario
#Descripción: Usando el diccionario creado en el ejercicio 2, actualiza el valor correspondiente a una clave existente. Ejemplo:
cursos.update({"Power BI": ["Sandra", "José", "Marta", "Pedro"]})
print(cursos) 
cursos["Power BI"] += ["Carlos"]
print(cursos)
#Eliminar un par clave-valor de un diccionario
cursos.pop("Tableau")
print(cursos)

```

```python
#Actualizar el valor de una clave en un diccionario
#Descripción: Usando el diccionario creado en el ejercicio 2, actualiza el valor correspondiente a una clave existente. Ejemplo:
cursos.update({"Power BI": ["Sandra", "José", "Marta", "Pedro"]})
print(cursos) 
cursos["Power BI"] += ["Carlos"]
print(cursos)
#Eliminar un par clave-valor de un diccionario
#Descripción: Usando el diccionario creado en el ejercicio 2, elimina un par clave-valor específico que indique la usuaria (debereis usar la función input() para preguntar qué clave quiere borrar).

eliminar_curso = input("Indica un curso para eliminarlo: ")
if eliminar_curso in cursos:
    cursos.pop(eliminar_curso)
    print(cursos)
else:
    print("El curso indicado no existe en el diccionario.")

#Verificar si una clave está presente en un diccionario
#Descripción: Usando el diccionario creado en el ejercicio 2, verifica si una clave específica está presente en él. La clave a buscar debe ser pasada por la usuaria, deberéis usar un input().
consultar_cursos = input("Indica un curso para saber si se dicta y los alumnos inscriptos: ")
if consultar_cursos in cursos:    
    print(f"El curso {consultar_cursos} tiene los siguientes alumnos inscriptos: {cursos[consultar_cursos]}")
else:
    cursos_disponibles = list(cursos.keys())
    print(
        f"El curso {consultar_cursos} no se encuentra disponible, seleccione alguno de la lista: {', '.join(cursos_disponibles[:-1])} y {cursos_disponibles[-1]}.")
#Obtener una lista de todas las claves de un diccionario
#Descripción: Usando el diccionario creado en el ejercicio 2, obtén una lista con todas las claves que contiene.
print(list(cursos.keys()))
print(*cursos.keys(), sep='\n') #uno por renglón
#Obtener una lista de todos los valores de un diccionario
#Descripción: Usando el diccionario creado en el ejercicio 2, obtén una lista con todos los valores que contiene.
print(list(cursos.values()))
print(*cursos.values(), sep='\n') #uno por renglón
#Utilizar el método setdefault() para establecer un valor predeterminado si la clave no existe
#Descripción: Usando el diccionario creado en el ejercicio 2, obten el valor correspondiente a una clave. Si la clave no existe, establece un valor predeterminado y devuelve dicho valor.
cursos.setdefault("JAVA", ["sin alumnos inscriptos al momento"])
print(cursos)
#Manipulación de un diccionario utilizando pop() y popitem()}
#Descripción: Usando el diccionario creado en el ejercicio 2, realiza operaciones de eliminación utilizando pop() y popitem(), y observa cómo afectan al diccionario. 
cursos.pop("SQL")
print(cursos)
cursos.popitem() #elimina el último elemento agregado, en este caso el curso JAVA.
print(cursos)
```
