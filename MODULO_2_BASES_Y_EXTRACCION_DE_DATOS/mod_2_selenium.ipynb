```python
# Importar librerías para tratamiento de datos
# -----------------------------------------------------------------------
import pandas as pd  # Pandas es una poderosa librería para manipulación y análisis de datos en Python.

# Importar librerías para web scraping y manipulación de datos
# -----------------------------------------------------------------------
from bs4 import BeautifulSoup
import requests

# Importar librerías para automatización de navegadores web con Selenium
# -----------------------------------------------------------------------
from selenium import webdriver  # Selenium es una herramienta para automatizar la interacción con navegadores web.
from webdriver_manager.chrome import ChromeDriverManager  # ChromeDriverManager gestiona la instalación del controlador de Chrome.
from selenium.webdriver.common.keys import Keys  # Keys es útil para simular eventos de teclado en Selenium.
from selenium.webdriver.support.ui import Select  # Select se utiliza para interactuar con elementos <select> en páginas web.

# Importar librerías para pausar la ejecución
# -----------------------------------------------------------------------
from time import sleep  # Sleep se utiliza para pausar la ejecución del programa por un número de segundos.

# Configuraciones
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None)  # Establece una opción de Pandas para mostrar todas las columnas de un DataFrame.
```

```python
#Inicializar el navegador web Chrome, acceder a la url y maximinar la ventana.
driver = webdriver.Chrome()
url="https://www.decathlon.es/"
driver.get(url)
print("Se accedió exitosamente a la web.")
driver.maximize_window()
# Esperar 5 segundos para que la página cargue completamente
sleep(5)
# Aceptar las cookies haciendo clic en el botón correspondiente
driver.find_element("css selector", "#didomi-notice-agree-button").click()
print("Aceptadas las cookies")
# Esperar 2 segundos después de aceptar las cookies
sleep(2)
```

```python
# Solicitar input de búsuqeda al usuario y utilizar el buscador del sitio

busqueda=input("¿Qué quisieras buscar?")
sleep(3)
driver.find_element("css selector", "#search-input-header-modal").send_keys(busqueda,Keys.ENTER)
sleep(20)
```

```python
diccionario_resultados= {"nombre": [], "precio":[], "puntuación":[]}
```

```python
for i in range (0,5):
    sleep(20)
    driver.find_element("css selector", f"#plp-{i}-title").click()
    sleep(20)
    nombre=driver.find_element("css selector", "#content > article > section.product-view__buy-box.product-view__buy-box--sticky-to-bottom > div > div.product-info__name > h1").text
    diccionario_resultados['nombre'].append(nombre)
    sleep(20)
    puntuación=driver.find_element("css selector", '#content > article > section.product-view__buy-box.product-view__buy-box--sticky-to-bottom > div > div.product-info__review > div > div.review__rating > div > p').text
    diccionario_resultados['puntuación'].append(puntuación)
    sleep(20)
    precio=driver.find_element("css selector",'#content > article > section.product-view__buy-box.product-view__buy-box--sticky-to-bottom > form > section.buy-box__section.buy-box__section--price > div.vp-price.vp-price--large > span').text
    diccionario_resultados['precio'].append(precio)
    driver.back()
    sleep(20)
```
