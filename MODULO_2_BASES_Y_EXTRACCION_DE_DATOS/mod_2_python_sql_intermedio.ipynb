```python
#Librerías
import mysql.connector
from mysql.connector import errorcode
from sqlalchemy import create_engine,FLOAT, VARCHAR, INTEGER, DATE, SmallInteger
from sqlalchemy.sql.sqltypes import String
import pandas as pd
```

```python
# Establecer conexión con MySQL
try:
    cnx=mysql.connector.connect(user='root', password='my-secret-pw', host ='127.0.0.1', database= 'tienda')
    print("Se estableció la conexión con Mysql.")
except mysql.connector.Error as err:
    print(f"No se pudo establecer la conexión. Error: {err}")
```

```python
# Acceso a la Base de Datos 'tienda' y consulta de nombre de columnas.

try:
    mycursor=cnx.cursor()
    query_tienda='USE tienda'
    mycursor.execute(query_tienda)
    print("Se accedió a la Base de Datos 'tienda'.")
    query_info_col="SELECT COLUMN_NAME FROM INFORMATION_SCHEMA.COLUMNS WHERE TABLE_SCHEMA = 'tienda' AND TABLE_NAME = 'customers'"
    mycursor.execute(query_info_col)
    resultado=[col[0] for col in mycursor.fetchall()]
    print(resultado)
        
except mysql.connector.Error as err:
    print(f"No se pudo acceder a la tabla. Error: {err}")
        
```

```python
# Consulta de customers en ciudad de Londres o Seattle
query_1="SELECT * FROM customers WHERE city IN ('London','Seattle')"
mycursor.execute(query_1)
resultado_q1=mycursor.fetchall()
df=pd.DataFrame(resultado_q1, columns=resultado)
df
```

```python
# Consulta de customers
query_2="SELECT * FROM customers"
mycursor.execute(query_2)
resultado_q2=mycursor.fetchall()
df2=pd.DataFrame(resultado_q2, columns=resultado)
df2.sample(15)
```
