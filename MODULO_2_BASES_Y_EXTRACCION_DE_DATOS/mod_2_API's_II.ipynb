### CONSULTA PRIMEROS 200 REGISTROS DE JUGADORES

```python
def consulta_jugadores():

    jugadores=[]
    counter=0

    while len(jugadores)<200:
        try:
            respuesta =  requests.get(url, headers = {"Authorization": apikey},params = {'cursor':counter,'per_page':100})
            respuesta.status_code
            
            if respuesta.status_code != 200:
                print(f"Se produjo el error: {respuesta.status_code}")
                return None
            
            ball_players=respuesta.json()
            jugadores.extend(ball_players['data']) #me traigo toda la info dsps filtrar
                 
            counter=ball_players['meta'].get('next_cursor')
            if counter is None:
                break
        except:
            break
    return jugadores[:200]
```

```python
datos_200_jugadores=consulta_jugadores()    #se almacena en variable el resultado de todos los datos para los 200 jugadores
```

### CONSULTA CON TIME PARA EVITAR "ERROR 429" AL CONSULTAR JUGADORES (TOTAL)

```python
import time 
def consulta_jugadores_total():

    jugadores=[]
    counter=0

    while True:
        try:
            respuesta =  requests.get(url, headers = {"Authorization": apikey},params = {'cursor':counter,'per_page':100})
            respuesta.status_code
                
            if respuesta.status_code != 200:
                print(f"Se produjo el error: {respuesta.status_code}")
                return None
                
            ball_players=respuesta.json()
            jugadores.extend(ball_players['data']) #me traigo toda la info dsps filtrar
                 
            counter=ball_players['meta'].get('next_cursor')
            if counter is None:
                break

            time.sleep(25)

        except:
            break

    return jugadores



```

## RED DATA API

```python
url= "https://apidatos.ree.es/es/datos/generacion/estructura-generacion"
cabeceras = {
    'Accept': 'application/json',
    'Content-Type': 'application/json',
    'Host': 'apidatos.ree.es'
}
parametros = {
    'start_date':'2021-01-01T00:00',
    'end_date':'2022-12-31T23:59',
    'time_trunc':'month',
    'geo_limit':'nacional'
    }

def consultar_electric_system():
    respuesta=requests.get(url, headers=cabeceras,params=parametros)
    respuesta.status_code

    if respuesta.status_code !=200:
        print(f"No se pudo realizar la consulta, error: {respuesta.status_code}.")
    else:
        datos=respuesta.json()
        datos_e=datos['included']
        print(json.dumps(datos_e, indent=2, ensure_ascii=False))
        return datos_e
```

```python
energia=consultar_electric_system()
```

lista de diccionarios:'data','meta','included'
lista de diccionarios de tipos de energia --> 'included': 
1.key 'type'
2.key 'attributes' + key 'type' esta el valor (renovable/norenovable) 
3.key 'attributes' + key 'values' :lista de diccionarios (12) m + keys 'datetime'
4.key 'attributes' + key 'values' lista de diccionarios (12) m + keys 'value'
