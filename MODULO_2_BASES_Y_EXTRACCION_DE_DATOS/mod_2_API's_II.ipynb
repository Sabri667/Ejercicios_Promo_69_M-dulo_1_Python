```python
import requests
import pandas as pd
import time
import json


apikey="146f6a53-7b10-49bf-ad60-3b916f299ced"
url="https://api.balldontlie.io/v1/players"
```

### CONSULTA PRIMEROS 200 REGISTROS DE JUGADORES

```python
def consulta_jugadores():

    jugadores=[]
    counter=0

    while len(jugadores)<200:
        try:
            respuesta =  requests.get(url, headers = {"Authorization": apikey},params = {'cursor':counter,'per_page':100})
            respuesta.status_code
            
            if respuesta.status_code != 200:
                print(f"Se produjo el error: {respuesta.status_code}")
                return None
            
            ball_players=respuesta.json()
            jugadores.extend(ball_players['data']) #me traigo toda la info dsps filtrar
                 
            counter=ball_players['meta'].get('next_cursor')
            if counter is None:
                break
        except:
            break
    return jugadores[:200]
```

```python
datos_200_jugadores=consulta_jugadores()    #se almacena en variable el resultado de todos los datos para los 200 jugadores
```

### CONSULTA CON TIME PARA EVITAR "ERROR 429" AL CONSULTAR JUGADORES (TOTAL)

```python
import time 
def consulta_jugadores_total():

    jugadores=[]
    counter=0

    while True:
        try:
            respuesta =  requests.get(url, headers = {"Authorization": apikey},params = {'cursor':counter,'per_page':100})
            respuesta.status_code
                
            if respuesta.status_code != 200:
                print(f"Se produjo el error: {respuesta.status_code}")
                return None
                
            ball_players=respuesta.json()
            jugadores.extend(ball_players['data']) #me traigo toda la info dsps filtrar
                 
            counter=ball_players['meta'].get('next_cursor')
            if counter is None:
                break

            time.sleep(25)

        except:
            break

    return jugadores



```
