#### Consulta de tabla y conversión con Pandas.

```python

try:
# Establecer conexción con MySQL.
    cnx= mysql.connector.connect(user='root', password='my-secret-pw', host='127.0.0.1')
    print("Conexión establecida con éxito")
except mysql.connector.Error as err:
    print(f"Error al conectar con MySQL: {err}")
# Posicionarse en la Base de Datos.
try:
    mycursor=cnx.cursor()
    mycursor.execute('USE alumnas_promo_69')
    print("Ya estamos en la BD 'alumnas_promo_69'.")
except mysql.connector.Error as err:
    print(f"No se pudo acceder a la Base de Datos: {err}")
# Realizar consulta de la tabla.
try:
    query_datos_2= ('SELECT * FROM alumnas')
    mycursor.execute(query_datos_2)
    resultado=mycursor.fetchall()
    print ("Se realizó la consulta a la tabla 'alumnas' y se almacenó en 'resultado'.")
except mysql.connector.Error as err:
    print(f"Error al acceder a la tabla: {err}")
```

```python
df=pd.DataFrame(resultado, columns= ['legajo','nombre','apellidos','ciudad','telefono'])
df
```

```python
#Consultar tabla de alumnas con nombre que contengan letra e y no sean de Madrid.
try:
    query_datos_3= "SELECT * FROM alumnas WHERE nombre LIKE '%e%' AND ciudad <> 'Madrid'"
    mycursor.execute(query_datos_3)
    resultado_3=mycursor.fetchall()
except mysql.connector.Error as err:
    print(f"Se produjo un error: {err}")
#Consultar id_alumnas y apellido de alumnas de la ciudad de Valencia.
try:
    query_datos_4= "SELECT id_alumna,apellidos FROM alumnas WHERE ciudad = 'Valencia'"
    mycursor.execute(query_datos_4)
    resultado_4=mycursor.fetchall()
except mysql.connector.Error as err:
    print(f"Se produjo un error: {err}")
```
