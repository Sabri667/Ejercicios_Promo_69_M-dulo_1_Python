```python
import requests

def buscar_film(nombre):
    
    url = f"https://swapi.py4e.com/api/films/?search={nombre}"
    respuesta = requests.get(url)
    
    if respuesta.status_code == 200:
        datos_film = respuesta.json()
        return datos_film['results']
    else:
        print(f"Error en la consulta: {respuesta.status_code}")
        return None

#consulta a usuario y llamado de búsqueda
pelicula_busqueda = input("Ingresa el nombre (o parte) de una película de Star Wars: ")
pelicula = buscar_film(pelicula_busqueda)     #devuelve None o lista de diccionarios(primeros 10).

#mensaje de resultado
if pelicula:
    print(f"\nSe encontraron {len(pelicula)} película(s):\n")
    for i,p in enumerate(pelicula,start=1):
        print(f"{i}. Nombre: {p['title']} | Director: {p['director']} | Año:{p['release_date']}")
else:
    print("No se encontraron películas con ese nombre.")
```

```python
import requests
#extraccion de datos - total
def buscar_personajes(nombre):
    """Busca personajes y devuelve una lista con TODOS los resultados."""
    url = f"https://swapi.py4e.com/api/people/?search={nombre}"
    todos_los_personajes = []
    total_resultados = 0

    while url:
        respuesta = requests.get(url)

        if respuesta.status_code == 200:
            datos = respuesta.json()
            #conteo total de la primera página
            if total_resultados == 0:
                total_resultados = datos['count']
            #se añade personajes de la página actual a la lista
            todos_los_personajes.extend(datos['results'])
            #obtenemos la URL de la siguiente página
            url = datos['next']
        else:
            print(f"Error en la consulta: {respuesta.status_code}")
            return None

    print(f"Total de personajes encontrados para la búsqueda: {total_resultados}")
    return todos_los_personajes

# consulta de búsqueda
nombre_busqueda = input("Ingresa el nombre (o parte) de un personaje de Star Wars: ")
personajes = buscar_personajes(nombre_busqueda)

if personajes:
    print(f"\nTu búsqueda coincide con {len(personajes)} personaje(s):\n")
    for i,p in (enumerate(personajes,start=1)):
        print(f"{i}.Nombre: {p['name']} | Género: {p['gender']}")
else:
    print("No se encontraron personajes con ese nombre.")
```

```python
import requests

#PERSONAJES: extraccion de datos total
def buscar_personajes(nombre):
    """Busca personajes y devuelve una lista con TODOS los resultados."""
    url = f"https://swapi.py4e.com/api/people/?search={nombre}"
    todos_los_personajes = []
    total_resultados = 0

    while url:
        respuesta = requests.get(url)

        if respuesta.status_code == 200:
            datos = respuesta.json()
            #conteo total de la primera página
            if total_resultados == 0:
                total_resultados = datos['count']
            #se añade personajes de la página actual a la lista
            todos_los_personajes.extend(datos['results'])
            #obtenemos la URL de la siguiente página
            url = datos['next']
        else:
            print(f"Error en la consulta: {respuesta.status_code}")
            return None

    return todos_los_personajes

# consulta de búsqueda
nombre_busqueda = input("Ingresa el nombre (o parte) de un personaje de Star Wars: ")
personajes = buscar_personajes(nombre_busqueda)

if personajes:
    print(f"\nTu búsqueda coincide con {len(personajes)} personaje(s):\n")
    for i,p in (enumerate(personajes,start=1)):
        print(f"{i}.Nombre: {p['name']} | Género: {p['gender']}")
else:
    print("No se encontraron personajes con ese nombre.")
print("---------------xxx----------------")
###--------------------------------------------------------------------------------------------------------###

#PERSONAJES: extraccion de datos (primeros 10 resultados)
def buscar_personajes_max_10(nombre):
    url = f"https://swapi.py4e.com/api/people/?search={nombre}/"
    respuesta = requests.get(url)
    
    if respuesta.status_code == 200:
        datos = respuesta.json()
        return datos['results']
    else:
        print(f"Error en la consulta: {respuesta.status_code}")
        return None

#consulta a usuario y llamado de búsqueda
nombre_busqueda = input("Ingresa el nombre (o parte) de un personaje de Star Wars: ")
personajes = buscar_personajes_max_10(nombre_busqueda)     #devuelve None o lista de diccionarios(primeros 10).

#mensaje de resultado
if personajes:
    print(f"\nTu búsqueda coincide con {len(personajes)} personaje(s):\n")
    for p in personajes:
        print(f"Nombre: {p['name']} | Género: {p['gender']}")
else:
    print("No se encontraron personajes con ese nombre.")
print("---------------xxx----------------")
###--------------------------------------------------------------------------------------------------------###

#PELÍCULAS
def buscar_film(nombre):
    
    url = f"https://swapi.py4e.com/api/films/?search={nombre}"
    respuesta = requests.get(url)
    
    if respuesta.status_code == 200:
        datos_film = respuesta.json()
        return datos_film['results']
    else:
        print(f"Error en la consulta: {respuesta.status_code}")
        return None

#consulta a usuario y llamado de búsqueda
pelicula_busqueda = input("Ingresa el nombre (o parte) de una película de Star Wars: ")
pelicula = buscar_film(pelicula_busqueda)     #devuelve None o lista de diccionarios(primeros 10).

#mensaje de resultado
if pelicula:
    print(f"\nTu búsqueda coincide con {len(pelicula)} película(s):\n")
    for i,p in enumerate(pelicula,start=1):
        print(f"{i}. Nombre: {p['title']} | Director: {p['director']} | Año:{p['release_date']}")
else:
    print("No se encontraron películas con ese nombre.")
print(f"---------------xxx----------------")
###--------------------------------------------------------------------------------------------------------###

#PLANETA: extraccion de datos (primeros 10 resultados)
def buscar_planeta(nombre):
    url = f"https://swapi.py4e.com/api/planets/?search={nombre}"
    respuesta = requests.get(url)
    
    if respuesta.status_code == 200:
        datos_planeta = respuesta.json()
        return datos_planeta['results']
    else:
        print(f"Error en la consulta: {respuesta.status_code}")
        return None

#consulta a usuario y llamado de búsqueda
planeta_busqueda = input("Ingresa el nombre (o parte) de un planeta de Star Wars: ")
planeta = buscar_planeta(planeta_busqueda)     #devuelve None o lista de diccionarios(primeros 10).

#mensaje de resultado
if planeta:
    print(f"\nSe encontraron {len(planeta)} planeta(s):\n")
    for i,p in enumerate(planeta,start=1):
        #formatear población
        poblacion_raw = p.get('population', 'None')
        if poblacion_raw != 'unknown':
            try:
                poblacion_formateada = f"{int(poblacion_raw):,.0f}"
            except ValueError:
                poblacion_formateada = poblacion_raw  # por si no es número
        else:
            poblacion_formateada = "unknown"

        print(f"{i}. Nombre: {p['name']} | Clima: {p['climate']} | Poblacion:{poblacion_formateada}")
else:
    print("No se encontraron planetas con ese nombre.")
print("---------------xxx----------------")

###--------------------------------------------------------------------------------------------------------###
 
#NAVES
def buscar_nave(nombre):
    
    url = f"https://swapi.py4e.com/api/starships/?search={nombre}"
    respuesta = requests.get(url)
    
    if respuesta.status_code == 200:
        datos_film = respuesta.json()
        return datos_film['results']
    else:
        print(f"Error en la consulta: {respuesta.status_code}")
        return None

#consulta a usuario y llamado de búsqueda
nave_busqueda = input("Ingresa el nombre (o parte) de una nave de Star Wars: ")
nave = buscar_nave(nave_busqueda)     #devuelve None o lista de diccionarios(primeros 10).

#mensaje de resultado
if nave:
    print(f"\nSe encontraron {len(nave)} nave(s):\n")
    for i,p in enumerate(nave,start=1):
        print(f"{i}. Nombre: {p['name']} | Modelo: {p['model']} | Fabricante:{p['manufacturer']}")
else:
    print("No se encontraron naves con ese nombre.")
print("---------------xxx----------------")


```
