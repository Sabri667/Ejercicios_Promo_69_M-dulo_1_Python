```python
#Librerías
import pandas as pd
import mysql.connector
from mysql.connector import errorcode
import csv
```

```python
# Establecer conexión con MySQL
try:
    cnx=mysql.connector.connect(user='root', password='my-secret-pw', host ='127.0.0.1', database= 'videoclub')
    print("Se estableció la conexión con Mysql.")
except mysql.connector.Error as err:
    print(f"No se pudo establecer la conexión. Error: {err}")
```

```python
# Lectura de archivo .csv y generación de lista de values.

with open('tabla_clientes.csv', 'r') as f:
    datos = list(csv.reader(f))
```

```python
print(datos)
```

```python
# Creación de cursor e ingreso de  datos en tabla 'clientes'.
try:
    mycursor=cnx.cursor()
    query_videoclub='USE videoclub'
    mycursor.execute(query_videoclub)
    print("Se accedió a la Base de Datos 'videoclub'.")
    mycursor.executemany(f"INSERT INTO clientes(nombre,apellido,direccion,telefono,email,fecha_registro) VALUES ({','.join(['%s']*len(datos[0]))})", datos)
    cnx.commit()
        
except mysql.connector.Error as err:
    print(f"No se pudo acceder a la tabla. Error: {err}")
        
```

```python
# Se crea consulta a la tabla 'clientes' y se desconecta de mysql.
query_consulta_clientes='SELECT * FROM clientes'
mycursor.execute(query_consulta_clientes)
resultado=mycursor.fetchall()
print(resultado)
cnx.close()
```

```python
# Se crea un DataFrame de la ultima consulta: 'clientes'.
df=pd.DataFrame(resultado,columns=('id_cliente','nombre','apellido','direccion','teléfono','email','fecha_registro'))
df
```
