```python
#Itera en la siguiente lista de temperaturas para crear tres listas nuevas.
#temperaturas = [2,45,67,23, 45,66,12,1,34,67,99,98,67,64, 44, 100]
#En cada una de las listas nuevas, tendremos temperaturas que cumplan con las siguientes condiciones:
#En la primera tengamos temperaturas entre 0 - 10, llama a la nueva lista bajas_temp.
#En la segunda tengamos temperaturas entre 11- 20, llama a la nueva lista medias_temp.
#En la tercera tengamos temperaturas por encima de 20, llama a la nueva lista altas_temp
temperaturas = [2,45,67,23,45,66,12,1,34,67,99,98,67,64,44,100]

bajas_temp=[] 
altas_temp=[]
medias_temp=[]

for i in temperaturas:
    if i > 20:
        altas_temp.append(i)
    elif 10<= i <=20:
        medias_temp.append(i)
    else:
        bajas_temp.append(i)

print(f"bajas_temp= {bajas_temp}")
print(f"medias_temp= {medias_temp}")
print(f"altas_temp= {altas_temp}")
#Las temperaturas están en grados centígrados. Para cada una de las listas creadas en el ejercicio anterior, convertid las temperaturas a grados fahrenheit 
#la relacion de grados celsius y fahrenheit es:
#(0 °C × 9 / 5) + 32 = 32 
#Por ejemplo, para convertir 2 grados centígrados a fahrenheit, el resultado sería:
#(2 °C × 9 / 5) + 32 = 35.6 °F
for i in bajas_temp:
    for i in range(len(bajas_temp)):
        bajas_temp[i] = round((bajas_temp[i] * (9/5)) + 32, 1)
print (bajas_temp)

for i in altas_temp:
    for i in range(len(altas_temp)):
        altas_temp[i] = round((altas_temp[i] * (9/5)) + 32,1)
print (altas_temp)

for i in medias_temp:
    for i in range(len(medias_temp)):
        medias_temp[i] = round((medias_temp[i] * (9/5)) + 32,1)
print (medias_temp)

#Imprime los números del 0 al 5 (ambos incluidos) usando range().
for i in range(0,6):
    print(i)

print("---")  # separador


#Imprime los números los números del 3 al 6 (ambos incluidos) usando range().
for i in range(0,7):
    if i in (3,4,5,6,7):
        print(i)

print("---")  # separador


#Imprime los números 6, 7 y 8 usando range().
for i in range(0,10):
    if i == 6 or i == 7 or i == 8:
        print(i)

print("---")  # separador


#Imprime los números 2, 3 y 4 usando range().
for i in range(2,5):
    print (i)

print("---")  # separador


#Imprime un rango que empiece en 2, aumentando cada paso de 3 en 3, que pare antes de llegar al número 13.
for i in range(2,13,3):
    print(i)


print("---")  # separador


#Imprime un rango con todos los valores impares que hay hasta 10.
for i in range(10):
    if i % 2 != 0:
        print(i)


```

```python
#Crea una lista nueva donde tengamos los números de la lista temperaturas elevados al cuadrado.
temp_cuadrado=[]
for i in temperaturas:
    temp_cuadrado.append(i**2)
print(temp_cuadrado)
print("---")  # separa
#Crea una lista nueva donde tengamos el resto de la división de cada uno de los números entre 2.
temp_resto=[]
for i in temperaturas:
    temp_resto.append(i%2)
print(temp_resto)
print("---")  # separa
#Crea una lista nueva donde tengamos solo los elementos pares de la lista temperaturas.
temp_elementos_pares=temperaturas[::2]
print(temp_elementos_pares)
print("---")  # separa
#Crea una lista nueva donde tengamos solo los elementos que sean menores que 50.
temp_menores_50=[]
for i in temperaturas:
    if i < 50:
        temp_menores_50.append(i)
print(f"Se generó la lista con temperaturas menores de 50 grados: {temp_menores_50}")
print("---")  # separa
#Crea una lista donde tengamos los valores "par" e "impar" en función de los valores de la lista original. La lista resultante tiene que ser como la que indicamos a continuación:
lista_strings=[]
for i in temperaturas:
    if i % 2 == 0:
        lista_strings.append("par")
    else:
        lista_strings.append("impar")
print(f"Se generó una lista de strings detallando si las temperaturas son o no pares: {lista_strings}.")
print("---")  # separa
#Pide al usuario un número entero positivo y suma los números pares e impares que hay de 0 hasta ese número en dos listas diferentes. Por ejemplo, si nos dan el número 4, la lista de pares sería [0, 2, 4] y la lista de impares sería [1, 3].

while True:
    n=int(input("Por favor, ingresa un número entero positivo: "))
    lista_par=[]
    lista_impar=[]
    if n >=0:
        for i in range(n+1):
            if i%2 == 0:
                lista_par.append(i)
            else:
                lista_impar.append(i)
        print(f"Seleccionaste el número '{n}' y se generó la lista par:{lista_par} y la lista impar: {lista_impar}.")
        break
    else:
        print("Por favor indica un número entero positivo o pulsa 'Esc' para cerrar el programa.")
print("---")  # separa

#Pide al usuario ingresar una cadena y cuenta cuántas veces aparece cada vocal (a, e, i, o, u) en la cadena.
frase=input("Por favor, ingresa una frase: ").lower()
contador_vocales={v:0 for v in "aeiou"}

for i in frase:
    if i in "aeiou":
        contador_vocales[i]+=1

print(f"Tu frase '{frase}' tiene el siguiente número de vocales: {contador_vocales}")
print("---")  # separa
```
